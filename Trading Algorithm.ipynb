```python
prices['daily_change']=((prices['close']-prices['open'])/prices['open'])*100
```

```python
prices['date']=pd.to_datetime(prices['date'])
```

```python
prices['rise']=(prices.close > prices.open)
```

```python
prices.drop(columns=[''],inplace=True)
```

```python
prices.set_index(['date','symbol'],inplace=True)
```

```python
agga={
    'open':['min','max']}                    
priceagg=prices.groupby(['symbol',prices.date.dt.year,prices.date.dt.month]).agg(agga)
```

```python
prices['prevprevday_daily_change']=(prices['daily_change'].groupby(level=1).shift(2))
```

```python
prices['2day_rise']=(prices.yday_daily_change >0) & (prices.prevprevday_daily_change>0)
```
